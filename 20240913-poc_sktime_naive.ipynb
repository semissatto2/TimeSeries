```python
from sktime.utils.plotting import plot_series
```

```python
from sktime.forecasting.naive import NaiveForecaster
```

```python
import sktime
print("sktime version: " + sktime.__version__)
import pandas as pd
print("pandas version: " + pd.__version__)
# print np version
import numpy as np
print("numpy version: " + np.__version__)
```

```python
def generate_time_series(num_days):
    # Create a date index starting from '2010-01-01'
    dates = pd.date_range(start='2010-01-01', periods=num_days, freq='D')
    
    values = np.random.normal(loc=100, scale=15, size=num_days)  # loc is the mean, scale is the standard deviation
    
    time_series = pd.Series(data=values, index=dates)
    
    return time_series
```

```python
forecaster = NaiveForecaster(strategy="mean", window_length=28)

N = 365*10
y = generate_time_series(N)
y_pred = forecaster.fit_predict(y=y, fh=y.index)

# plotting predictions and past data
plot_series(y, y_pred, labels=["y", "y_pred"])
```

```python
# create simple rolling mean forecaster
y_pred_rolling = y.rolling(window=28).mean()

# plotting predictions and past data
plot_series(y, y_pred_rolling, labels=["y", "y_pred_rolling"])
```
